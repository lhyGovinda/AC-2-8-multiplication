```python
import numpy as np
from qiskit import *
from qiskit_aer import Aer
q1=QuantumRegister(24)
qc=QuantumCircuit(q1)
c1 = ClassicalRegister(8)

#linear layer,A1*psi
qc=QuantumCircuit(q1)
def A1psi(q1):
    qc.cx(1,5)
    qc.cx(2,1)
    qc.cx(4,2)
    qc.cx(5,7)
    qc.cx(5,0)
    qc.cx(7,1)
    qc.cx(2,5)
    qc.cx(0,4)
    qc.cx(6,2)
    qc.cx(3,7)
    qc.cx(2,3)
    qc.cx(7,6)
    qc.cx(4,7)
    qc.cx(1,4)
    qc.cx(4,2)
    qc.swap(0,2)
    qc.swap(0,4)
    qc.swap(0,6)
    qc.swap(0,7)
    qc.swap(1,3)
    qc.swap(1,5)

    qc.cx(9,13)
    qc.cx(10,9)
    qc.cx(12,10)
    qc.cx(13,15)
    qc.cx(13,8)
    qc.cx(15,9)
    qc.cx(10,13)
    qc.cx(8,12)
    qc.cx(14,10)
    qc.cx(11,15)
    qc.cx(10,11)
    qc.cx(15,14)
    qc.cx(12,15)
    qc.cx(9,12)
    qc.cx(12,10)
    qc.swap(8,10)
    qc.swap(8,12)
    qc.swap(8,14)
    qc.swap(8,15)
    qc.swap(9,11)
    qc.swap(9,13)


#linear layer,A2*A1_inv
qc=QuantumCircuit(q1)
def A2A1_inv(q1):
    qc.cx(2,4)
    qc.cx(1,2)
    qc.cx(6,1)
    qc.cx(1,3)
    qc.cx(7,2)
    qc.cx(4,6)
    qc.cx(3,7)
    qc.cx(5,1)
    qc.cx(0,5)
    qc.cx(4,3)
    qc.cx(7,0)
    qc.cx(1,7)
    qc.cx(2,5)
    qc.cx(5,4)
    qc.swap(0,7)
    qc.swap(0,1)
    qc.swap(0,2)
    qc.swap(0,6)
    qc.swap(3,4)

    qc.cx(10,12)
    qc.cx(9,10)
    qc.cx(14,9)
    qc.cx(9,11)
    qc.cx(15,10)
    qc.cx(12,14)
    qc.cx(11,15)
    qc.cx(13,9)
    qc.cx(8,13)
    qc.cx(12,11)
    qc.cx(15,8)
    qc.cx(9,15)
    qc.cx(10,13)
    qc.cx(13,12)
    qc.swap(8,15)
    qc.swap(8,9)
    qc.swap(8,10)
    qc.swap(8,14)
    qc.swap(11,12)

#linear layer, (A2*psi)_inv
qc=QuantumCircuit(q1)
def A2psi_inv(q1):
    qc.cx(6,4)
    qc.cx(4,5)
    qc.cx(6,2)
    qc.cx(1,4)
    qc.cx(0,6)
    qc.cx(6,3)
    qc.cx(7,0)
    qc.cx(3,7)
    qc.cx(1,3)
    qc.cx(5,7)
    qc.cx(2,1)
    qc.cx(4,6)
    qc.swap(1,4)
    qc.swap(1,3)
    qc.swap(1,6)
    qc.swap(1,7)
    qc.swap(1,2)

    qc.cx(14,12)
    qc.cx(12,13)
    qc.cx(14,10)
    qc.cx(9,12)
    qc.cx(8,14)
    qc.cx(14,11)
    qc.cx(15,8)
    qc.cx(11,15)
    qc.cx(9,11)
    qc.cx(13,15)
    qc.cx(10,9)
    qc.cx(12,14)
    qc.swap(9,12)
    qc.swap(9,11)
    qc.swap(9,14)
    qc.swap(9,15)
    qc.swap(9,10)

#linear layer, B1_inv*psi
qc=QuantumCircuit(q1)
def B1_invpsi(q1):
    qc.cx(18,16)
    qc.cx(16,21)
    qc.cx(21,23)
    qc.cx(20,16)
    qc.cx(17,20)
    qc.cx(19,18)
    qc.cx(23,17)
    qc.cx(22,19)
    qc.cx(21,19)
    qc.cx(16,23)
    qc.cx(20,18)
    qc.cx(19,20)
    qc.cx(18,21)
    qc.cx(18,22)
    qc.swap(16,17)
    qc.swap(16,18)
    qc.swap(19,22)
    qc.swap(19,23)


#linear layer,B2_inv*B1
qc=QuantumCircuit(q1)
def B2_invB1(q1):
    qc.cx(22,18)
    qc.cx(23,20)
    qc.cx(18,19)
    qc.cx(17,23)
    qc.cx(18,17)
    qc.cx(20,18)
    qc.cx(16,21)
    qc.cx(16,17)
    qc.cx(19,16)
    qc.cx(23,19)
    qc.cx(21,22)
    qc.cx(21,23)
    qc.cx(17,21)
    qc.cx(20,22)
    qc.cx(16,20)
    qc.cx(22,16)
    qc.swap(16,22)
    qc.swap(16,17)
    qc.swap(16,19)
    qc.swap(16,18)
    qc.swap(16,23)

#linear layer, psi_inv*B2
qc=QuantumCircuit(q1)
def psi_invB2(q1):
    qc.cx(21,23)
    qc.cx(23,16)
    qc.cx(22,23)
    qc.cx(17,22)
    qc.cx(22,21)
    qc.cx(21,19)
    qc.cx(20,22)
    qc.cx(16,17)
    qc.cx(19,20)
    qc.cx(19,16)
    qc.cx(22,18)
    qc.cx(16,22)
    qc.cx(22,19)
    qc.swap(16,21)
    qc.swap(16,23)
    qc.swap(16,17)
    qc.swap(16,19)
    qc.swap(16,22)
    qc.swap(18,20)

# 4 multiplication_h in F2^2
qc=QuantumCircuit(q1)
def mul_h(q1):
    qc.cx(17,16)
    qc.ccx(0,8,17)
    qc.ccx(1,9,16)
    qc.cx(0,1)
    qc.cx(8,9)
    qc.cx(17,16)
    qc.ccx(1,9,17)
    qc.cx(0,1)
    qc.cx(8,9)

    qc.cx(19,18)    
    qc.ccx(2,10,19)
    qc.ccx(3,11,18)
    qc.cx(2,3)
    qc.cx(10,11)
    qc.cx(19,18)
    qc.ccx(3,11,19)
    qc.cx(2,3)
    qc.cx(10,11)

    qc.cx(21,20)
    qc.ccx(4,12,21)
    qc.ccx(5,13,20)
    qc.cx(4,5)
    qc.cx(12,13)
    qc.cx(21,20)
    qc.ccx(5,13,21)
    qc.cx(4,5)
    qc.cx(12,13)

    qc.cx(23,22)
    qc.ccx(6,14,23)
    qc.ccx(7,15,22)
    qc.cx(6,7)
    qc.cx(14,15)
    qc.cx(23,22)
    qc.ccx(7,15,23)
    qc.cx(6,7)
    qc.cx(14,15)


#Logic diagram of quantum circuit in Fig. 9
qc=QuantumCircuit(q1,c1)

qc.x(4)
qc.x(12)
qc.x(16)
qc.x(17)
qc.x(19)
qc.x(20)   #initial state

A1psi(q1)
B1_invpsi(q1)

qc.barrier(q1)

mul_h(q1)

qc.barrier(q1)

A2A1_inv(q1)
B2_invB1(q1)

qc.barrier(q1)

mul_h(q1)

qc.barrier(q1)

A2psi_inv(q1)
psi_invB2(q1)

for i in range(8):
    qc.measure(q1[16+i],c1[i])

qc.draw('mpl',fold=36)
```

```python
#Simulation and validation
from qiskit_aer import Aer
simulator = Aer.get_backend('aer_simulator')
result = simulator.run(qc, shots=10, memory=True).result()
memory = result.get_memory(qc)
print(memory)
```
